```python
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
x = np.linspace(-20, 20, 200)
p_x = norm.pdf(x, 0, 5)
p_c_A = 1 - np.exp(-x**2/40)
plt.plot(x, p_c_A)
```

```python
p_c_B = 1/2*(1-p_x) + p_c_A*p_x
plt.plot(x, p_c_B)
```

```python
u_B = p_c_B*(1-p_c_B)
plt.plot(x, u_B) 
```
